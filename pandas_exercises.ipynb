## Filter, Sort, and Group by

Sort values by `"a"`in ascending order

```python
df.sort_values("a")
```

Sort values by "b" in descending order

```python
df.sort_values("b", ascending=False)
```

## Join/Combine

```python
df2 = pd.DataFrame({"a":[5,7,0,8,5], "b": [9,42,45,1,87]})
```

Add the rows in df to the end of df2

```python
df.append(df2)
```

Add the columns in df to the right end of df2 by columns and then by rows

```python
df3 = pd.concat([df, df2], axis=1) # 1 for columns, 0 for rows
df3
```

```python
df4 = pd.concat([df, df2], axis=0) # 0 for rows
df4
```
